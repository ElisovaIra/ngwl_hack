```python
from collections import Counter

# для катег фичей
def top1(s):
    return Counter(s).most_common()[0][0]

def top2(s):
    try:
        return Counter(s).most_common()[1][0]
    except:
        return np.nan

def top3(s):
    try:
        return Counter(s).most_common()[2][0]
    except:
        return np.nan
    
def top4(s):
    try:
        return Counter(s).most_common()[3][0]
    except:
        return np.nan    
    
def top5(s):
    try:
        return Counter(s).most_common()[4][0]
    except:
        return np.nan        
    
def len_set(s):
    return len(set(s))

# для числ фичей
func_list = [np.sum, np.mean, np.median, np.std, min, max, 
#                 lambda x: np.quantile(x, 0.95),
#                 lambda x: np.quantile(x, 0.05),
               ]
```

```python
grouped_shipments = shipments.groupby(['phone_id', 'order_month'], as_index=False).agg({ 
#     'total_cost': func_list,
#     'rate': func_list,
#     'promo_total': func_list,
#     'total_weight': func_list,
    'retailer': [top1, top2, top3, len_set],
    's.order_state': [top1, top2, len_set],
    'shipment_state': [top1, top2, len_set],
    's.city_name': [top1, top2, len_set], 
    's.store_id': [top1, top2, top3, top4, top5, len_set], 
    'dw_kind': [top1, top2, len_set], 
    'platform': [top1, top2, len_set], 
    'os': [top1, top2, len_set]
})

grouped_shipments.columns = ["_".join(x) for x in grouped_shipments.columns.ravel()]
grouped_shipments.rename(columns={'phone_id_': 'phone_id', 
                                  'order_month_': 'order_month'}, inplace=True)
```
